```python
from gurobipy import *
```

```python
import pandas as pd

basic_info = pd.read_csv('instance01.csv')
display(basic_info)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('max_colwidth', 100)
```

```python
import numpy as np

job_id = basic_info['Job ID']
jobs = len(job_id)
D = basic_info['Due Time']
stage1Time = basic_info['Stage-1 Processing Time']
stage2Time = basic_info['Stage-2 Processing Time']
p = [[0] * jobs for i in range(2)]
for i in range(jobs):
    p[0][i] = stage1Time[i]
for i in range(jobs):
    p[1][i] = stage2Time[i]

available1 = basic_info['Stage-1 Machines']
available2 = basic_info['Stage-2 Machines']
available2 = available2.fillna(0)

machines = 0
for i in range(jobs):
    for item in available1[i]:
        if(item.isdigit()):
            item = int(item)
            if(item > machines):
                machines = item

for i in range(jobs):
    if(available2[i] == 0):
        continue
    for item in available2[i]:
        if(item.isdigit()):
            item = int(item)
            if(item > machines):
                machines = item

available_machine = np.zeros((2,jobs,machines))

for i in range(jobs):
    for item in available1[i]:
        if(item.isdigit()):
            item = int(item)
            available_machine[0][i][item-1] = 1
for i in range(jobs):
    if(available2[i] == 0):
        continue
    for item in available2[i]:
        if(item.isdigit()):
            item = int(item)
            available_machine[1][i][item-1] = 1
```

```python
print(available_machine)
```

```python
eg1 = Model("eg1")

y = []
for i in range(2):
    y.append([])
    for j in range(jobs):
        y[i].append([])
        for k in range(machines):
            y[i][j].append(eg1.addVar(lb = 0, vtype = GRB.BINARY, name = "y" + str(i+1) + str(j+1) + str(k+1)))

c = []
for i in range(2):
    c.append([])
    for j in range(jobs):
        c[i].append(eg1.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = "c" + str(i+1) + str(j+1)))

T = []
for j in range(jobs):
    T.append(eg1.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = "D" + str(j+1)))

z = []
for i in range(2):
    z.append([])
    for j in range(jobs):
        z[i].append([])
        for a in range(2):
            z[i][j].append([])
            for b in range(jobs):
                z[i][j][a].append([])
                for k in range(machines):
                    z[i][j][a][b].append(eg1.addVar(lb = 0, vtype = GRB.BINARY, name = "z" + str(i+1) + str(j+1) + str(a+1) + str(b+1) + str(k+1)))
```

```python
eg1.setObjective(
    quicksum(T[j] for j in range(jobs)), GRB.MINIMIZE)

M = 0
for i in range(2):
    for j in range(jobs):
        M += p[i][j]

for j in range(jobs):
    eg1.addConstr(c[0][j] <= c[1][j] - p[1][j])
    eg1.addConstr(c[0][j] + 1 + p[1][j] >= c[1][j])
    eg1.addConstr(c[0][j] >= p[0][j])
    eg1.addConstr(T[j] >= c[1][j] - D[j])
    eg1.addConstr(T[j] >= 0)
for i in range(2):
    for j in range(jobs):
        out = True
        for a in range(machines):
            if(available_machine[i][j][a] == 1):
                out = False
                break
        if(out == True):
            continue
        eg1.addConstr(quicksum(y[i][j][k] for k in range(machines)) == 1)
for i in range(2):
    for j in range(jobs):
        for k in range(machines):
            eg1.addConstr(available_machine[i][j][k] - y[i][j][k] >= 0)

for i in range(2):
    for j in range(jobs):
        for a in range(2):
            for b in range(jobs):
                for k in range(machines):
                    if(i == a and j == b):
                        continue
                    eg1.addConstr(y[i][j][k] + y[a][b][k] <= (z[i][j][a][b][k] + z[a][b][i][j][k] + 1))
                    eg1.addConstr(c[i][j] + p[a][b] - c[a][b] <= M * (1 - z[i][j][a][b][k]))

eg1.optimize()  
```

```python
print(eg1 == GRB.OPTIMAL)
```

```python
print("total tardiness = ", eg1.objVal)

print("whether stage i of job j is processed in machine k:")
for j in range(jobs):
    for i in range(2):
        for k in range(machines):
            print(y[i][j][k].x, "\t", end = "")
        print("")
    print("")

print("the completion time of stage i of job j:")
for j in range(jobs):
    for i in range(2):
        print(c[i][j].x, "\t", end = "")
    print("")

print("the tadiness of job j:")
for j in range(jobs):
    print(T[j].x, "\t", end = "")

# print("whether stage i of job j is processed in front of stage i' of job j' in machine k:")
# for i in range(2):
#     for j in range(jobs):
#         for a in range(2):
#             for b in range(jobs):
#                 for k in range(machines):
#                     print("is stage " + str(i+1)  + " of job " + str(j+1) + " in front of stage " + str(a+1) + " job " + str(b+1) + " in machine " + str(k+1))
#                     print(z[i][j][a][b][k].x, "\t", end="")
#                     print()
```

```python
eg2 = Model("eg2")

y = []
for i in range(2):
    y.append([])
    for j in range(jobs):
        y[i].append([])
        for k in range(machines):
            y[i][j].append(eg2.addVar(lb = 0, vtype = GRB.BINARY, name = "y" + str(i+1) + str(j+1) + str(k+1)))

c = []
for i in range(2):
    c.append([])
    for j in range(jobs):
        c[i].append(eg2.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = "c" + str(i+1) + str(j+1)))

T = []
for j in range(jobs):
    T.append(eg2.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = "D" + str(j+1)))

z = []
for i in range(2):
    z.append([])
    for j in range(jobs):
        z[i].append([])
        for a in range(2):
            z[i][j].append([])
            for b in range(jobs):
                z[i][j][a].append([])
                for k in range(machines):
                    z[i][j][a][b].append(eg2.addVar(lb = 0, vtype = GRB.BINARY, name = "z" + str(i+1) + str(j+1) + str(a+1) + str(b+1) + str(k+1)))

w = eg2.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = "w")
```

```python
obj = eg1.objVal
print(obj)
```

```python
eg2.setObjective(w, GRB.MINIMIZE)

M = 0
for i in range(2):
    for j in range(jobs):
        M += p[i][j]

eg2.addConstrs(w >= c[1][j] for j in range(jobs))

for j in range(jobs):
    eg2.addConstr(c[0][j] <= c[1][j] - p[1][j])
    eg2.addConstr(c[0][j] + 1 + p[1][j] >= c[1][j])
    eg2.addConstr(c[0][j] >= p[0][j])
    eg2.addConstr(T[j] >= c[1][j] - D[j])
    eg2.addConstr(T[j] >= 0)

eg2.addConstr(quicksum(T[j] for j in range(jobs)) <= obj)

for i in range(2):
    for j in range(jobs):
        out = True
        for a in range(machines):
            if(available_machine[i][j][a] == 1):
                out = False
                break
        if(out == True):
            continue
        eg2.addConstr(quicksum(y[i][j][k] for k in range(machines)) == 1)
for i in range(2):
    for j in range(jobs):
        for k in range(machines):
            eg2.addConstr(available_machine[i][j][k] - y[i][j][k] >= 0)

for i in range(2):
    for j in range(jobs):
        for a in range(2):
            for b in range(jobs):
                for k in range(machines):
                    if(i == a and j == b):
                        continue
                    eg2.addConstr(y[i][j][k] + y[a][b][k] <= (z[i][j][a][b][k] + z[a][b][i][j][k] + 1))
                    eg2.addConstr(c[i][j] + p[a][b] - c[a][b] <= M * (1 - z[i][j][a][b][k]))
eg2.optimize()
```

```python
print(eg2 == GRB.OPTIMAL)
```
